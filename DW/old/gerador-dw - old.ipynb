```python
import pandas as pd
import numpy as np

# Definindo o número de linhas
n = 0

```

```python
# 1. Tabela Dimensão: Vendedores
n=10
vendedores = pd.DataFrame({
    'VendedorID': range(1, n+1),
    'Nome': [f'Vendedor {i}' for i in range(1, n+1)],
    'Região': np.random.choice(['Norte', 'Sul', 'Leste', 'Oeste'], n),
    'DataAdmissao': pd.to_datetime(np.random.choice(pd.date_range('2018-01-01', '2024-01-01', freq='D'), n)),
    'Status': np.random.choice(['Ativo', 'Inativo'], n),
})


vendedores.to_excel('vendedores.xlsx',index=False)
vendedores
```

```python
# 2. Tabela Dimensão: Clientes
n=50
clientes = pd.DataFrame({
    'ClienteID': range(1, n+1),
    'NomeCliente': [f'Cliente {i}' for i in range(1, n+1)],
    'Cidade': np.random.choice(['São Paulo', 'Rio de Janeiro', 'Belo Horizonte', 'Porto Alegre'], n),
    'DataCadastro': pd.to_datetime(np.random.choice(pd.date_range('2020-01-01', '2024-01-01', freq='D'), n)),
    'TipoCliente': np.random.choice(['Novo', 'Fidelizado'], n),
})
clientes.to_excel('clientes.xlsx',index=False)

clientes
```

```python
# 3. Tabela Dimensão: Produtos
n=4
produtos = pd.DataFrame({
    'ProdutoID': range(1, n+1),
    'NomeProduto': np.random.choice(['Produto A', 'Produto B', 'Produto C', 'Produto D'], n),
    'Categoria': np.random.choice(['Categoria 1', 'Categoria 2', 'Categoria 3'], n),
    'Preço': np.random.uniform(50, 5000, n).round(2),
})
produtos.to_excel('produtos.xlsx',index=False)

produtos
```

```python

# 4. Tabela Dimensão: Regiões
regioes = pd.DataFrame({
    'RegiaoID': range(1, 5),
    'NomeRegiao': ['Norte', 'Sul', 'Leste', 'Oeste'],
    'Populacao': np.random.randint(500000, 20000000, 4),  # População estimada por região
})
regioes.to_excel('regioes.xlsx',index=False)

regioes
```

```python
# Vamos supor que as tabelas de 'produtos', 'vendedores' e 'clientes' já foram criadas como no exemplo anterior

n = 6000  # Número de vendas

vendas = []

for i in range(1, n + 1):
    venda_id = i
    vendedor_id = np.random.choice(vendedores['VendedorID'])
    cliente_id = np.random.choice(clientes['ClienteID'])
    data_venda = pd.to_datetime(np.random.choice(pd.date_range('2023-01-01', '2024-12-31', freq='D')))
    
    # Selecionar entre 1 a 3 produtos para cada venda
    produtos_venda = np.random.choice(produtos['ProdutoID'], size=np.random.randint(1, 4), replace=True)
    
    # Gerar a quantidade de cada produto vendido
    quantidade_produtos = np.random.randint(1, 5, len(produtos_venda))
    
    # Obter os preços dos produtos vendidos
    valor_produtos = produtos.loc[produtos['ProdutoID'].isin(produtos_venda), 'Preço'].values
    
    # Verificar se as quantidades e preços têm o mesmo tamanho
    if len(quantidade_produtos) == len(valor_produtos):
        valor_venda = (quantidade_produtos * valor_produtos).sum()  # Total da venda
    else:
        # Caso haja discrepância (não esperado), continuamos o loop
        continue

    # Para cada produto da venda, criamos uma linha
    for j in range(len(produtos_venda)):
        vendas.append([venda_id, vendedor_id, cliente_id, produtos_venda[j], data_venda, valor_produtos[j], quantidade_produtos[j]])

# Criando o DataFrame de vendas
vendas_df = pd.DataFrame(vendas, columns=['VendaID', 'VendedorID', 'ClienteID', 'ProdutoID', 'DataVenda', 'ValorProduto', 'Quantidade'])

vendas_df.to_excel('vendas_df.xlsx',index=False)

vendas_df
```
